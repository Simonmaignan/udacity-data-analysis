# Act Report
We decided to analyze the archive from the [WeRateDogs](https://en.wikipedia.org/wiki/WeRateDogs) twitter account ([@dog_rates](https://twitter.com/dog_rates)).

After gathering the data from multiple sources (a given tweets archive as csv file, a tweet image predictions tsv file downloaded from the web and some additional information about the tweets using the Twitter API Tweepy), we assessed and cleaned it.  
From this wrangled data set, we were able to retrieve a few insights about the WeRateDogs tweets.  
Here is a succinct list of the insights:
- The cleaned data set contains 2075 tweets.
- Out of which, 73.8% contained a picture predicted to represent a dog.
- We computed a few simple statistics on the dog ratings for different set of tweets
  - With all the cleaned tweets, the rating mean is 10.65/10, the standard deviation is 39.83 and the median is 10/10
  - After removing 2 outliers, the rating mean is 10.60/10, the standard deviation drops to 2.16 (makes more sense) and the median is 10/10
  - With only tweets of predicted dogs, the rating mean is 10.95/10, the standard deviation is 1.68 and the median is 10/10 and the outliers are not there anymore.
- The scatter plot below of the favorite_count compared with the retweet_count shows us that there is a correlation between the 2 variables  

- The analysis of the given dog stage shows us that there are 15.0% of the tweets, predicted as dogs, were the user gave the stage of his dog. The pie chart below can tell us that most of the dogs have a pupper stage.
